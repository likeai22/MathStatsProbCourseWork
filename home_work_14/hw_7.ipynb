```python
import numpy as np
import scipy.stats as stats
```

## 1. Выбрать тест и проверить, есть ли различия между выборками: 

```python
# Данные
x1 = [380, 420, 290]
y1 = [140, 360, 200, 900]

# Применение U-критерия Манна-Уитни
statistic, p_value = stats.mannwhitneyu(x1, y1, alternative="two-sided")

# Вывод результатов
print(f"Статистика U: {statistic}")
print(f"P-значение: {p_value}")

# Проверка гипотезы
alpha = 0.05
if p_value < alpha:
    print(
        "Отвергаем нулевую гипотезу: есть статистически значимые различия между выборками."
    )
else:
    print(
        "Нет оснований отвергать нулевую гипотезу: различия между выборками не являются значимыми."
    )
```

```python
# Данные
x1 = [380, 420, 290]
y1 = [140, 360, 200, 900]

# Объединение данных и ранжирование
combined = x1 + y1
ranks = stats.rankdata(combined)  # Ранжируем объединённые данные

# Разделяем ранги обратно на две выборки
ranks_x1 = ranks[: len(x1)]  # Ранги для x1
ranks_y1 = ranks[len(x1) :]  # Ранги для y1

# Размеры выборок
n1 = len(x1)
n2 = len(y1)

# Суммы рангов
T1 = sum(ranks_x1)
T2 = sum(ranks_y1)

# Выбор максимальной выборки
n_max = max(n1, n2)
T_max = T2 if n_max == n2 else T1

# Вычисление U по формуле
U = n1 * n2 + (n_max * (n_max + 1)) / 2 - T_max

# Вывод результата
print(f"Статистика U: {U}")
```

## 2. Исследовалось влияние препарата на уровень давления пациентов.

```python
# Данные
x1 = [150, 160, 165, 145, 155]  # До приема препарата
x2 = [140, 155, 150, 130, 135]  # Через 10 минут
x3 = [130, 130, 120, 130, 125]  # Через 30 минут

# Применение теста Фридмана
statistic, p_value = stats.friedmanchisquare(x1, x2, x3)

# Вывод результатов
print(f"Статистика Фридмана: {statistic}")
print(f"P-значение: {p_value}")

# Проверка гипотезы
alpha = 0.05
if p_value < alpha:
    print(
        "Отвергаем нулевую гипотезу: есть статистически значимые различия между измерениями."
    )
else:
    print(
        "Нет оснований отвергать нулевую гипотезу: различия между измерениями не являются значимыми."
    )
```

## 3. Сравните 1 и 2 е измерения, предполагая, что 3го измерения через 30 минут не было.

```python
# Данные
x1 = [150, 160, 165, 145, 155]  # До приема препарата
x2 = [140, 155, 150, 130, 135]  # Через 10 минут

# Применение критерия Вилкоксона
statistic, p_value = stats.wilcoxon(x1, x2)

# Вывод результатов
print(f"Статистика Вилкоксона: {statistic}")
print(f"P-значение: {p_value}")

# Проверка гипотезы
alpha = 0.05
if p_value < alpha:
    print(
        "Отвергаем нулевую гипотезу: есть статистически значимые различия между измерениями."
    )
else:
    print(
        "Нет оснований отвергать нулевую гипотезу: различия между измерениями не являются значимыми."
    )
```

## 4. Даны 3 группы  учеников плавания.

```python
# Данные
x1 = [56, 60, 62, 55, 71, 67, 59, 58, 64, 67]  # Группа 1
x2 = [57, 58, 69, 48, 72, 70, 68, 71, 50, 53]  # Группа 2
x3 = [57, 67, 49, 48, 47, 55, 66, 51, 54]  # Группа 3

# Применение критерия Краскела-Уоллиса
statistic, p_value = stats.kruskal(x1, x2, x3)

# Вывод результатов
print(f"Статистика Краскела-Уоллиса: {statistic}")
print(f"P-значение: {p_value}")

# Проверка гипотезы
alpha = 0.05
if p_value < alpha:
    print(
        "Отвергаем нулевую гипотезу: есть статистически значимые различия между группами."
    )
else:
    print(
        "Нет оснований отвергать нулевую гипотезу: различия между группами не являются значимыми."
    )
```

## 5. Заявляется, что партия изготавливается со средним арифметическим 2,5 см. 

Проверить данную гипотезу, если известно, что размеры изделий подчинены нормальному закону распределения.  
Объем выборки 10, уровень статистической значимости 5%  

2.51, 2.35, 2.74, 2.56, 2.40, 2.36, 2.65, 2.7, 2.67, 2.34

```python
n = 10
m = 2.5
x1 = np.array([2.51, 2.35, 2.74, 2.56, 2.40, 2.36, 2.65, 2.7, 2.67, 2.34])

mean = x1.mean()
std = x1.std(ddof=1)
t_fact = (mean - m) / std * np.sqrt(n)

t_o = stats.t.ppf(0.95, 9)
t_t = stats.t.ppf(0.975, 9)

print(f"mean: {mean:.3f}")
print(f"std: {std:.3f}")
print(f"t-факт: {t_fact:.3f}")
print(f"t-тест односторонний = {t_o:.3f}")
print(f"t-тест двусторонний = {t_t:.3f}")
```
